```python
import pandas as pd
import random
```

```python
# Define 20 countries and their regions
country_region_mapping = {
    "United States": "North America",
    "Canada": "North America",
    "Australia": "Oceania",
    "Venezuela": "South America",
    "Spain": "Southern Europe",
    "Brazil": "South America",
    "Cape Verde": "Western Africa",
    "India": "Southern Asia",
    "South Africa": "Southern Africa",
    "Mexico": "Central America",
    "France": "Western Europe",
    "Germany": "Western Europe",
    "Japan": "Eastern Asia",
    "China": "Eastern Asia",
    "Italy": "Southern Europe",
    "Argentina": "South America",
    "New Zealand": "Oceania",
    "Egypt": "Northern Africa",
    "Thailand": "South-Eastern Asia",
    "Russia": "Eastern Europe"
}

# Function to generate dummy data
def generate_dummy_data(num_entries):
    data = []
    countries = list(country_region_mapping.keys())
    
    for _ in range(num_entries):
        # Randomly select current and next countries
        current_country = random.choice(countries)
        next_country = random.choice(countries)
        
        # Get the corresponding regions
        current_region = country_region_mapping[current_country]
        next_region = country_region_mapping[next_country]
        
        # Append the entry to the data list
        data.append([current_country, current_region, next_country, next_region])
    
    # Create a DataFrame
    columns = ["current_country", "current_region", "next_country", "next_region"]
    df = pd.DataFrame(data, columns=columns)
    
    return df

# Generate 1000 entries
dummy_data = generate_dummy_data(1000)
```
